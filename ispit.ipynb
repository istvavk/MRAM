```python
import numpy as np
import scipy as sp
import sympy as sym
import matplotlib.pyplot as plt
import numpy.linalg as npLa
import scipy.linalg as spLa
```

```python
n = 5
A = np.zeros((n, n))
first_row = np.array([-8, -28, -58, -67, -30])
A[0] = first_row
A[np.arange(1, n), np.arange(0, n-1)] = 1
B = np.ones((5, 1))
B[0] = 2
B[1] = -1
B[3] = -1
C = np.array([[1, 1, 0, 1, 0], 
              [2, 0, 0, 1, 1]])
D = 0
E = np.eye(n)
```

```python
P = spLa.solve_continuous_lyapunov(A, -B@B.T)
Q = spLa.solve_continuous_lyapunov(A.T, -C.T@C)
R = npLa.cholesky(P)
L = npLa.cholesky(Q)

M = L.T @ R
[U, sigma, V_T] = npLa.svd(M)
V = np.conj(V_T.T)

sigma_sqrt = np.sqrt(sigma)
sigma_sqrt = 1 / sigma_sqrt
sigma_inv_sqrt = np.diag(sigma_sqrt)

T = R @ V @ sigma_inv_sqrt
T_inv = sigma_inv_sqrt @ U.T @ L.T

A_t = T_inv @ A @ T
B_t = T_inv @ B
C_t = C @ T
D_t = D

r = 2

A_r = A_t[:r, :r]
B_r = B_t[:r, :]
C_r = C_t[:, :r]
D_r = D_t

print("A_r iznosi: \n", A_r)
print("B_r iznosi: \n", B_r)
print("C_r iznosi: \n", C_r)
print("D_r iznosi: \n", D_r)

# c)

print(f"Ocjena pogreske: {2*sum(sigma[r:])}")
```

```python
# ZADATAK 2

# točke su s1 = 0.1 j i s2 = 1.5 j

print(f"Dimenzije nase matrice prijenosa su: {(C@(E-A)@B).shape}")

# imamo dva retka i jedan stupac, dakle b će nam biti skalar 1, a c će nam biti [1, 0] jer biramo tang vektore 
# tako da je interpolacija u prvom elementu

b = 1
c = np.array([[1],
              [0]])


# odredimo sigma i tau

sigma_1 = np.array([0.1])*1j
sigma_2 = np.array([1.5])*1j
tau_1 = np.array([0.1])*1j
tau_2 = np.array([1.5])*1j

v1 = npLa.solve(sigma_1[0]*E-A, np.eye(n))@B*b
v2 = npLa.solve(sigma_2[0]*E-A, np.eye(n))@B*b
w1 = npLa.solve(tau_1[0]*E.T-A.T, np.eye(n))@C.T@c
w2 = npLa.solve(tau_2[0]*E.T-A.T, np.eye(n))@C.T@c

V_r = np.block([v1, v2])
W_r = np.block([w1, w2])

E_r = W_r.T @ E @ V_r
A_r = W_r.T @ A @ V_r
B_r = W_r.T @ B
C_r = C @ V_r
D_r = D

print("E_r iznosi: \n", E_r)
print("A_r iznosi: \n", A_r)
print("B_r iznosi: \n", B_r)
print("C_r iznosi: \n", C_r)
print("D_r iznosi: \n", D_r)

```

```python
# b)

def trans_func(s, E, A, B, C, D):
    return C@npLa.solve(s*E-A, np.eye(np.size(A, 0)))@B + D

print(trans_func(sigma_2[0], E, A, B, C, D))
print(trans_func(sigma_2[0], E_r, A_r, B_r, C_r, D_r))

# zelimo izracunati (G-Gr)(1.5i)

print("(G-Gr)(1.5i)")
print(trans_func(sigma_2[0], E, A, B, C, D) - trans_func(sigma_2[0], E_r, A_r, B_r, C_r, D_r))
```
